# Summarize by Station

```python
df_stations = pd.read_csv('../data/02-prepped/stations-2023-02-to-2024-01.csv')
```

```python
df_stations['pct_january']=100*12*df_stations['rides_january']/df_stations['rides_all']
df_stations['pct_jan1416']=100*12*df_stations['rides_jan14_16']/df_stations['rides_all']
```

```python
df_stations
```

### filter dataset to include only frequently-used stations

```python
df_stations_large = df_stations[df_stations['rides_all']>=1000]
```

### top stations

```python
df_stations_large.sort_values('pct_january',ascending=False)
```

```python
df_stations_large.sort_values('pct_jan1416',ascending=False)
```

```python
df_stations_large.to_csv("../results/flourish-stations-january.csv", index=False)
```
